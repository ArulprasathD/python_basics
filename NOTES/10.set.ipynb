# add

```python
myset = {"apple", "banana", "cherry"}

myset.add("hyhy")

print(myset)
```

```python
myset = {"apple", "banana", "cherry"}
myset1 = {"pineapple", "mango", "papaya"}

myset.update(myset1)

print(myset)
```

```python
myset = {"apple", "banana", "cherry"}
mylist = {"kiwi", "orange","apple"}

myset.update(mylist)

print(myset)
```

# remove

```python
myset = {"apple", "banana", "cherry"}

myset.remove("banana")

print(myset)
```

```python
myset = {"apple", "banana", "cherry"}

myset.discard("banana")

print(myset)
```
