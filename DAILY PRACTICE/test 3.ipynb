```python
import math

rad = float(input("Enter radius of the circle : "))

area  = math.pi * rad ** 2

print(f"Area of circle : {area}")
```

```python
def check(num):
    if num >= 1:
        print(num," is positive")
        
    elif num == 0:
        print(num," is Zero")

    else:
        print(num," is Negative")
while True:
    
        user = input("Enter the number : ")
        if user.lower() == "exit":
            print("program ended")
            break
        try:
            num = int(user)
            check(num)
        except ValueError:
            print("Please enter a valid number")

```

```python
script = "hello,;"" world!!"

remove_punctuations = ""

for i in script:
    if i not in ".,?!:;""'-":
        remove_punctuations += i

print(remove_punctuations)
```

```python
import string

script = "hello,;"" World!!"

remove_punctuations = ""

for i in script:
    if i not in string.punctuation:
        remove_punctuations += i

print(remove_punctuations)
```
