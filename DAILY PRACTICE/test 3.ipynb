```python
def check(num):
    if num >= 1:
        print(num," is positive")
        
    elif num == 0:
        print(num," is Zero")

    else:
        print(num," is Negative")
while True:
    
        user = input("Enter the number : ")
        if user.lower() == "exit":
            print("program ended")
            break
        try:
            num = int(user)
            check(num)
        except ValueError:
            print("Please enter a valid number")

```

```python
script = "hello,;"" world!!"

remove_punctuations = ""

for i in script:
    if i not in ".,?!:;""'-":
        remove_punctuations += i

print(remove_punctuations)
```

```python
import string

script = "hello,;"" World!!"

remove_punctuations = ""

for i in script:
    if i not in string.punctuation:
        remove_punctuations += i

print(remove_punctuations)
```

```python
num = int(input("Enter decimal number: "))

binary = ""

while num > 0:
    rem = num % 2
    binary = str(rem) + binary
    num = num // 2

print("Binary:", binary)
```
