```python
class parent:
    def public_class(self):
        print("I 'm public")

    def _protected_class(self):
        print("I 'm protected")

    def __private_class(self):
        print("I 'm private")

    def call_class(self):
        print("SAME CLASS")

        self.public_class()
        self._protected_class()
        self.__private_class()

class child(parent):
    def child_class(self):
        self.public_class()
        self._protected_class()
        try:
            self.__private_class()
        except Exception as e:
            print(e)


p = parent()
c = child()

print("\n PARENT CLASS")
p.call_class()


print("\n CHILD CLASS")
c.child_class()
```
