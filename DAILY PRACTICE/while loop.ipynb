```python
#sum of digits

i = 9875

total = 0

while i > 0:
    digit = i%10
    total += digit
    i = i//10
    
print("sum of digit :",total)
```

```python
#reverse

i = 1234

reverse = 0

while i > 0:
    digit = i % 10
    reverse = reverse*10 + digit
    i = i // 10
print(reverse)
```

```python
#largest number 

i = 97564
lar = 0

while i > 0:
    digit = i % 10
    if lar < digit:
        lar = digit
    i = i // 10

print(lar)
```

```python
#second largest number

i = 2564
lar = 0
lar2 = 0

while i > 0:
    digit = i % 10
    if lar < digit:
        lar2 = lar
        lar = digit
    elif lar2 < digit:
        lar2 = digit
            
    i = i // 10

print(lar2)
```

```python
#count odd digit

i = 583246

count = 0

while i > 0:
    digit = i % 10
    if digit % 2 != 0:
        count += 1
    i = i//10
print(count)
```
