```python
class google_acc:
    def __init__(self,name,acc_no):
        self.name = name
        self.account = acc_no


    def login(self):
        return f"name : {self.name} / saved password : {self.account}"

name = input("Enter Name : ")
acc_no = input("Enter the Account No : ")

s1 = google_acc(name,acc_no)
print(s1.login())
```

```python
class mobile:

    def __init__(self,brand,price):
        self.brand = brand
        self.price = price

    def details(self):
        print(f"Brand : {self.brand}")
        print(f"price : {self.price}")

brand = input("Enter the Mobile Brand : ")
price = int(input("Enter the price : "))

j = mobile(brand,price)
j.details()
```

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello I am {self.name} and I am {self.age} years old"

name = input("Enter Name : ")
age = input("Enter Age : ")

s1 = person(name,age)
print(s1.greet())
```

```python
class laptop:


    
    def __init__(self,brand,ram):
        self.brand = brand
        self.ram = ram

    def spec(self):      
        if self.brand == "dell" and self.ram == 6:
            return "6gb Ram","512 SSd","price : 350000"
        elif self.brand == "acer" and self.ram == 6:
            return "6gb Ram","512 SSd","price : 320000"
        elif self.brand == "hp" and self.ram == 6:
            return "16gb Ram","512 SSd","price : 1350000"
        else:
            return "out of stock"

print("IN STOCK : DELL / HP / ACER")

brand = input("Enter laptop name :").lower()
ram = int(input("Enter ram : "))

s1 = laptop(brand,ram)
print(s1.spec())
        
```
