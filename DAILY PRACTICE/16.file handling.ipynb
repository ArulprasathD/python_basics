##  MODE

 r - read (file must exist)

w - write( overwrite and create)

a - append (add to end of the file)

r+ - read and write (over write and create)

w+ - write and read (file must exist)

a+ - append and read

rb -read binary

wb -write binary

ab -append binary

```python
file = open("notes.txt","w")                 #write
file.write("welcome to python file!\n")
file.write("topic is file handling\n")
file.close()
```

```python
file = open("notes.txt","r")
content = file.read()
print("content \n"content)
file.close()
```

```python
file = open("notes.txt","a")             #append
file.write("added new line \n")
file.close()
```

```python
with open("notes.txt","r")as file:                #with block
    for line in file:
        print(line.strip())
```

```python
feedback = input("Enter the feedback : ")

with open("feedback_log.txt","a") as log:
    log.write(feedback + "\n")

print("Thank you for your feedback !!")
```

```python
with open("file.txt","r")as f:
    while True:
        line = f.readline()
        if not line:
            break
        if "error" in line:
            print("Error found : ",line)
```
