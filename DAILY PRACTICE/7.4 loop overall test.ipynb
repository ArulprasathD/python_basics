```python
#anagram

str1 = input("Enter the word 1: ").lower()
str2 = input("Enter the word 2: ").lower()

count = 0
for i in str1:
    if i in str2:
        count += 1

if count == len(str1):
    print("anagram")
else:
    print("not anagram")
```

```python
#palindrome

a = "mon"

b = ''

for i in range(len(a)-1,-1,-1):
    b += a[i]

if b == a:
    print("palindrome")
else:
    print("not")
```
