```python
class instagram:
    def __init__(self):
        self.username = "aruuulllll / pubilc"
        self.__password = "a1r2u3l4 / private"
        self._email = "arxxxxx@gmail.com / protected"

    def details(self):
        print("user details")
        print("USERNAME : ",self.username)
        print("PASSWORD : ",self.__password)
        print("EMAIL :",self._email)

class users(instagram):
    def profile(self):
        print("profile visit")
        print("USERNAME : ",self.username)
        print("EMAIL :",self._email)
        try:
            print("PASSWORD : ",self.__password)
        except AttributeError:
            print("NO access")

class other:
    def varrrr(self):
        pass
        
j = instagram()
j1 = users()
j2 = other()

print("\n >> details <<")

j.details()

print("\n >> profile <<")

j1.profile()

```

```python
class access:
    def __init__(self):
        self.public_var = "I 'm public"
        self._protected_var = "I 'm protected"
        self.__private_var = "I 'm private"

    def from_main_class(self):
        print("From main class")
        print("public :",self.public_var)
        print("protected :",self._protected_var)
        print("private : ",self.__private_var)

class access1(access):
    def from_sub_class(self):
        print("From sub class")
        print("public : ",self.public_var)
        print("protected  : ",self._protected_var)
        try:
            print("private : ",self.__private_var)
        except:
            print("attribute error")

class outside():
    def other1(self,obj):
        print("unrelated class")
        print("public : ",obj.public_var)
        print("protected : ",obj._protected_var)
        try:
            print("private : ",obj.__private_var)
        except:
            print("no access")


c = access()
c1 = access1()
k = outside()


print("\n MAIN CLASS😏 ")
c.from_main_class()


print("\n SUB CLASS😒 ")
c1.from_sub_class()


print("\n OTHER CLASS🙄 ")
k.other1(c)
```
