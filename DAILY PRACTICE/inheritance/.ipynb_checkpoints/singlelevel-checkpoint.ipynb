```python
class dad:
    def house(self):
        return "I am from dad class"

class son(dad):
    def tv_shop(self):
        return "I am from son class"

s = son()
print(s.tv_shop())

print(s.house())
```

```python
class app1:
    def ver1(self):
        return "zomato version 0.01"

class app1_1(app1):
    def ver2(self):
        return "zomato version 0.11"

a = app1_1()
print("New version : ",a.ver2())
print("existing version : ",a.ver1()) 
```

```python
class app1:
    def ver1(self):
        return "zomato version 0.01"

class app1_1(app1):
    def ver2(self):
        return "zomato version 0.11"

    def var1(self):
        print("New update in version1 : zomato version 0.001")

a = app1_1()
print(a.ver1())
print("New version : ",a.ver2())
a.var1()
```
