```python
class grandfather:      #parent
    def house(self):
        print("grandpah : I have a 3BHK house")

class father(grandfather):        #child
    def shop(self):
        print("father : I have a Rice mill")

class son(father):                #child class
    def car(self):
        print("son : I have a lambo")


s1 = son()
s1.car()
s1.house()
s1.shop()
```

```python
class animal:
    def eat(self):
        print("Animal is eating")

class dog(animal):
    def bark(self):
        print("Dog is barking")

class babydog(dog):
    def weep(self):
        print("Baby dog is weeping")

a1 = babydog()
a1.eat()
a1.bark()
a1.weep()
```

```python
class Person:
    def __init__(self,name):
        self.name = name

    def display_name(self):
        print(f"Name : {self.name}")


class Employee(Person):
    def __init__(self,name,emp_id):
        super().__init__(name)
        self.emp_id = emp_id

    def display_id(self):
        print(f"Employee ID : {self.emp_id}")


class Manager(Employee):
    def __init__(self, name , emp_id , department):
        super().__init__(name, emp_id)
        self.department = department

    def display_department(self):
        print(f"Department : {self.department}")


name = input("Enter your name : ").capitalize()
emp_id = int(input("Enter your ID : "))
department = input("Enter your department : ").capitalize()

m1 = Manager(name,emp_id,department)

m1.display_name()
m1.display_id()
m1.display_department()
```
