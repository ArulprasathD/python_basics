```python
class dad : 
    def house(self):
        print("white house")

class mom:
    def shop(self):
        print("orange blend")

class son(dad,mom):
    def factory(self):
        print("Rubber factory")


j1 = son()
j1.house()
j1.shop()
j1.factory()
```

```python
class sub1:
    def __init__(self,eng,tam):
        self.eng = eng
        self.tam = tam

    def tam_mark(self):
        print("Tamil mark : ",self.tam)

class sub2:
    def eng_mark (self):
        print("English mark : ",self.eng)

class total_mark(sub1,sub2):
    def total (self):
        print("total : ",self.tam + self.eng)

        
eng = int(input("enter english mark : "))
tam = int(input("enter tamil mark : "))


j = total_mark(eng,tam)
j.tam_mark()
j.eng_mark()
j.total()   
```
