```python
num = int(input("Enter the number : "))

temp = str(num)

reverse=""

for i in range(len(temp)-1,-1,-1):
    reverse = reverse + temp[i]

print(reverse)
```

```python
num = 121

temp = num

reverse = 0


while num > 0:
    digit = num % 10
    reverse = reverse*10+digit

    num = num // 10

print(reverse)

if reverse == temp:
    print("palindrome")

else:
    print("not")
```

```python
num = "212"

reverse = ""

for i in range(len(num)-1,-1,-1):
    reverse += num[i]

print(reverse)

if reverse == num:
    print("palindrome")

else:
    print("Not")
```
