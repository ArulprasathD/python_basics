```python
num = int(input("Enter number : "))

power = len(str(num))
store = 0

for i in str(num):
    j = int(i) ** power
    store += j

print(store)

if store == num :
    print("Armstrong number")

else:
    print("no Armstrong number")
```

```python
def armstrong(num):

    power = len(str(num))
    store = 0
    
    for i in str(num):

        store += int(i) ** power
        

        
    if store == num:
        print("Armstrong",store)

    else:
        print("no")

        
num = int(input("Enter number"))

armstrong(num)
```
