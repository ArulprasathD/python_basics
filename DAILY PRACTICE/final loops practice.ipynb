```python
num = 121

temp = num

reverse = 0


while num > 0:
    digit = num % 10
    reverse = reverse*10+digit

    num = num // 10

print(reverse)

if reverse == temp:
    print("palindrome")

else:
    print("not")
```

```python
num = 1234567

even_digit_count = 0

while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even_digit_count = even_digit_count + 1

    num = num // 10

print("EVEN COUNT : ",even_digit_count)

```

```python
num = 1946583

highest_digit = 0

while num > 0:
    digit = num % 10

    if digit > highest_digit:
        highest_digit = digit

    num = num // 10

print("HIGHEST DIGIT IN SEQUENCE : ",highest_digit)
```
