```python
def check(num):
    if num >= 1:
        print(num," is positive")
        
    elif num == 0:
        print(num," is Zero")

    else:
        print(num," is Negative")
while True:
    
        user = input("Enter the number : ")
        if user.lower() == "exit":
            print("program ended")
            break
        try:
            num = int(user)
            check(num)
        except ValueError:
            print("Please enter a valid number")

```

```python
num = int(input("Enter decimal number: "))

binary = ""

while num > 0:
    rem = num % 2
    binary = str(rem) + binary
    num = num // 2

print("Binary:", binary)
```

```python
var =input("Enter the string : ").lower()

var1 = ""

for i in var:
    if i  not in var1:
        var1 += i

print(var1)

if var1 == var :
    print("UNIQUE")

else:
    print("NOT UNIQUE")
```

```python
class unique:
    
    def __init__(self,var):
        self.var = var

    def check(self):
        var1 = ""
        for i in self.var:
            if i not in var1:
                var1 += i

        if var1 == self.var:
            print("UNIQUE")
        else:
            print("NOT UNIQUE")

u = unique("abc")
u.check()


        
```

```python
my_list = [1,2,3]

store = []

for i in range(len(my_list)-1,-1,-1):
    store.append(my_list[i])

if store == my_list:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
class palinrome:
    def __init__(self,my_list):
        self.list = my_list

    def checker(self):
        reverse = []

        for i in range(len(self.list)-1,-1,-1):
            reverse.append(self.list[i])

        if reverse == self.list:
            print("palindrome")
        else:
            print("not")

p = palinrome([1,2,1])

p.checker()
```
