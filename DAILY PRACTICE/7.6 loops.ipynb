```python
#armstrong number

num = int(input("Enter the number : "))

store = 0

digit = len(str(num))

for i in str(num):
    store += int(i)**digit

#print(store)

if store == num:
    print("armstrong")

else:
    print("Not")
```

```python
for num in range(1, 1001):

    digits = len(str(num))
    total = 0

    for i in str(num):
        total += int(i) ** digits

    if total == num:
        print(num)
```

```python
a = [10,20,4,45,99]

lar = a[0]
sec = a[0]

for i in a:
    if i > lar:
        lar = i

for j in a:
    if j > sec and j != lar:
        sec = j

print("second large number : ",sec)
```

```python
num = int(input("Enter number : "))

if num <= 1:
    print("not prime")
else:
    for i in range(2,num):
        if num % i == 0:
            print("Not prime")
            break
    else:
        print("prime")
```
