```python
#anagram

word1 = input("Enter word 1 : ").lower()
word2 = input("Enter word 2 : ").lower()

count = 0

for i in word1:
    if i in word2:
        count += 1

if count == len(word1):
    print("Anagram")
else:
    print("no Anagram")
```

```python
#palindrome

char = input("Enter word : ")

store= ''

if char == char[::-1]:
    print("palindrome")
else:
    print("no palindrome")


for i in range(len(char)-1,-1,-1):
    store = store + char[i]


if store == char:
    print("palindrome")
else:
    print("not palindrome|")
```

```python
#prime

val = int(input("Enter value : "))

if val <= 1:
    print("not prime")
else:
    for i in range(2,val):
        if val % i == 0:
            print("not prime")
            break
        else:
            print("prime")
```

```python
# fibonacci series

a = 3
b = 4

for i in range(7):
    print(a)
    c = a + b
    a = b
    b = c
```

```python
num = int(input("Enter number: "))

temp = num
count = 0

# Step 1: count digits
while temp > 0:
    count += 1
    temp = temp // 10

temp = num
total = 0

# Step 2: calculate Armstrong sum
while temp > 0:
    digit = temp % 10
    total = total + digit ** count
    temp = temp // 10

# Step 3: check result
if total == num:
    print("Armstrong number")
else:
    print("Not Armstrong")
```
