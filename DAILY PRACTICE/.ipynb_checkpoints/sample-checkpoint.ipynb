```python
#armstrong number

num = int(input("Enter the number : "))

store = 0

digit = len(str(num))

for i in str(num):
    store += int(i)**digit

#print(store)

if store == num:
    print("armstrong")

else:
    print("Not")
```

```python
for num in range(1, 1001):

    digits = len(str(num))
    total = 0

    for i in str(num):
        total += int(i) ** digits

    if total == num:
        print(num)
```
