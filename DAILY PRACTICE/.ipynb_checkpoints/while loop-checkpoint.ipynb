```python
#sum of digits

i = 9875

total = 0

while i > 0:
    digit = i%10
    total += digit
    i = i//10
    
print("sum of digit :",total)
```

```python
#reverse

i = 1234

reverse = 0

while i > 0:
    digit = i % 10
    reverse = reverse*10 + digit
    i = i // 10
print(reverse)
```

```python
#largest number 

i = 97564
lar = 0

while i > 0:
    digit = i % 10
    if lar < digit:
        lar = digit
    i = i // 10

print(lar)
```

```python
#second largest number

i = 2564
lar = 0
lar2 = 0

while i > 0:
    digit = i % 10
    if lar < digit:
        lar2 = lar
        lar = digit
    elif lar2 < digit:
        lar2 = digit
            
    i = i // 10

print(lar2)
```

```python
#count odd digit

i = 583246

count = 0

while i > 0:
    digit = i % 10
    if digit % 2 != 0:
        count += 1
    i = i//10
print(count)
```

```python
i = 4589

rev = 0

while i > 0:
    val = i % 10
    rev = rev*10 + val
    i = i // 10

print(rev)
```

```python
i = 57249

even = 0
odd = 0

while i > 0:
    val = i % 10
    if val % 2 == 0:
        even += 1
    else:
        odd += 1
    i = i // 10
        
print("even : ",even)
print("odd  : ",odd)
```

```python
num = 234

product = 1

while num > 0:
    val = num % 10
    product *= val
    num = num // 10

print(product)
```

```python
i = 123

a = i
check = 0

while i > 0:
    val = i % 10
    check = check*10+ val
    i = i // 10

if a == check:
    print("palindrome")
else:
    print("not palindrome")
```

```python
i = 58321

lar = 0

while i > 0:
    val = i % 10
    if lar < val:
        lar = val
    i = i // 10

print(lar)
```

```python
i = 58321

lar = 0
lar2 = 0

while i > 0:
    val = i % 10
    if lar < val:
        lar2 = lar
        lar = val
    elif lar2 < val and val != lar:
            lar2 = val
    i = i // 10

print(lar)
print(lar2)
```

```python
i = 103010

total = 0

while i > 0:
    val = i % 10
    if val == 0:
        total += 1
    i = i // 10

print("count : ",total)
```
