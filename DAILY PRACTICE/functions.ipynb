```python
def checker(a):
    if a % 2 == 0:
        return "Even"
    else:
        return "Odd"

a = int(input("Enter number : "))
print(a, "is", checker(a))
```

```python
def fact(a):

    result = 1
    for i in range(1,a+1):
        result *= i
    return result

a = int(input("Enter number : "))
result = fact(a)
print("Factorial is ", result)
```

```python
def checker(num):

    if num <= 1:
        return "Not prime"
    else:
        for i in range(2,num):
            if num % i == 0:
                return "Not Prime"
                break
        else:
            return "Prime"

num = int(input("Enter number : "))
print(checker(num))
```

```python
def test(num):
    if num > 1:
        return "Positive"
    elif num == 0:
        return "Zero"
    else:
        return "Negative"
        
num = int(input("Enter number : "))

print(test(num))
```

```python
def palindrome(num):

    num_str = str(num)
    rev = ''

    for i in range(len(num_str)-1,-1,-1):
        rev = rev + num_str[i]

    if rev == num_str:
        return "Palindrome"
    else:
        return "Not Palindrome"

num = int(input("Enter the number : "))
print(palindrome(num))
```
