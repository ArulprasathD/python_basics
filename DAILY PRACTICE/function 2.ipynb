```python
def even(num):

    even_count = 0
    
    for i in str(num):
        if int(i) % 2 == 0:
            even_count += 1

    return even_count

num = int(input("Enter the number : "))
print("Even count : ",even(num))
```

```python
def reverse(num):

    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev*10+digit
        num = num // 10

    return rev

num = int(input("Enter number (more than 5 numbers) : "))
print("reverse : ",reverse(num))
```
