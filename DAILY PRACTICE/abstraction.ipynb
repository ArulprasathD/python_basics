```python
from abc import ABC,abstractmethod


class upcoming_update(ABC):

    @abstractmethod
    def optimize_third_party (self):
        pass

    @abstractmethod
    def show(self):
        #show after update msg
        pass

class update14_2(upcoming_update):
    def version(self):
        print("PD2165EF_EX_A_14.2.22.2.W30")

    def optimize_third_party(self):
        print("Optimized third-party app compatibility to fix the occasional issue where some apps do not work properly")

    def show(self):
        print("This upgrade will not delete or modify your usage data,")

version_call = update14_2()
version_call.version()
version_call.optimize_third_party()
version_call.show()
```
