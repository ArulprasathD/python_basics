```python
#3

nums = int(input("Enter number : "))

if nums <= 1:
    print("not prime ! ")
else:
    for i in range(2,nums):
        if nums % i == 0:
            print("not prime")
            break
    else:
            print("prime")
```

```python
# 1

num = int(input("Enter number : "))
convert = str(num)
total = 0


for i in convert:
    digit = int(i)
    k = digit ** 3
    total += k 

    
if total == num:
    print(total ,"==", num,"armstrong number")
else:
    print("no armstrong number")
```

```python
#2

num = int(input("Enter number : "))
convert = str(num)
store = ""

for i in range(len(convert)-1,-1,-1):
    store =store + convert[i]
print(store)

if store == convert:
    print("palindrome")
else:
    print("not palindrome")
    
    
```
