```python
#2

num = int(input("Enter number : "))

convert = str(num)
total_digit = 0

for i in convert:
    k = int(i)
    total_digit = total_digit + k
print(total_digit)
```

```python
#3

nums = int(input("Enter number : "))

if nums <= 1:
    print("not prime ! ")
else:
    for i in range(2,nums):
        if nums % i == 0:
            print("not prime")
            break
    else:
            print("prime")
```

```python
#4

var = input("Enter word: ")

reverse = ""

for i in range(len(var) - 1, -1, -1):
    reverse += var[i]

print("reverse : ",reverse)
     
```

```python
#5

l = ["a","b","a","c","b"]

after_remove = []

for i in l:
    if i not in after_remove:
        after_remove.append(i)
print(after_remove)
```
