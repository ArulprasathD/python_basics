```python
class Student:             #creating class
    def __init__(self,name,age,mark):
        self.name = name
        self.age = age
        self.mark = mark


    def details(self):
        return self.name,self.age,self.mark
            
name = input("Enter name : ")
age = int(input("Enter age : "))
mark = int(input("Enter mark : "))

s1=Student(name,age,mark)
print(s1.details())
```

```python
class car:
    def __init__(self,brand,model,price):
        self.brand = brand
        self.model = model
        self.price = price

    def show_car_details(self):
        print("BRAND : ",self.brand)
        print("MODEL : ",self.model)
        print("PRICE : ",self.price)

brand = input("Enter car brand : ").upper()
model = int(input("Enter car model : "))
price = input("Enter the price : ")

c1 = car(brand,model,price)

print("\n output")
c1.show_car_details()
```

```python
#single inheritance

class institue:
    def besant(self):
        print("head office,velachery")

class sub_institue(institue):
    def besant_1(self):
        print("branch,tambaram")

s1 = sub_institue()
s1.besant()
s1.besant_1()
```

```python
class parent:
    def house(self):
        print("i 'm from parent class")

class son(parent):
    def shop(self):
        print("i 'm from child class")

s1 = son()
s1.house()
s1.shop()
```

```python
class parent:
    def house(self):
        print("red")

class son(parent):
    def shop(self):
        print("yellow")

    def house(self):
        print("white")

s1 = son()
s1.house()
s1.shop()
```

```python
class grandpa:
    def house(self):
        print("white garden")

class dad(grandpa):
    def factory(self):
        print("oil factory")

class son(dad):
    def car(self):
        print("lambo")

s1= son()
s1.house()
s1.factory()
s1.car()
```

```python
class dad:
    def house (self):
        print("dad house")

class mom:
    def shop(self):
        print("mom shop")

class son(dad,mom):
    def factory(self):
        print("son factory !!")

a1 = son()
a1.factory()
a1.shop()
a1.house()           #multiple inheritance
```

```python
class HOD:
    def head(self):
        print("i 'm the head of the department")

class staff(HOD):
    def teach(self):
        print("i 'm the staff of AIDS department")

class student(HOD):
    def learn(self):
        print("i 'm the student of AIDS department")
s2 = staff()
s2.head()
s2.teach()

print("\n")

s1 = student()
s1.head()
s1.learn()
```

```python
#polymorphism


class dad:
    def house(self):
        print("house white")

class son(dad):
    def factory(self):
        print("gas factory")

    def house(self):                #over riding ✅
                                    #python don't have overloading ✖️
        print("house yellow")


s1 =son()
s1.house()
s1.factory()
```
