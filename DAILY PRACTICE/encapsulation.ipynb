```python
class zepto_order:
    def __init__(self,customer_name,order_item,discount,total_bill):
        self.name = customer_name
        self.order = order_item
        self.__discount = discount
        self.__total = total_bill

    def __final_bill(self):
        return self.__total - self.__discount

    def _admin_view(self):
        return {
            "Customer Name ":self.name,
            "Order Item ":self.order,
            "Discount  ":self.__discount,
            "Total Bill ":self.__final_bill()
        }

    def customer_view(self):
        return {
            "Customer Name ":self.name,
            "Order Item" :self.order,
            "Total Bill":self.__final_bill()
        }

class adminportal:
    def admin_view(self,zepto_order):
        return zepto_order._admin_view()

class customerapp:
    def user_view(self,zepto_order):
        return zepto_order.customer_view()

order = zepto_order("arul",["pizza","burger"],140,700)

admin = adminportal()
customer =customerapp()

print(admin.admin_view(order))

print(customer.user_view(order))
```

```python
class  food_order:
    def __init__(self,name,item,delivery_charge,total_amount):
        self.customer_name = name
        self.items = item
        self.__delivery_charge = delivery_charge         #private
        self.__total_amount = total_amount               #private

    def __final_amount(self):                                    #private
        return self.__total_amount + self.__delivery_charge

    def _admin_view(self):                                    #protected
        return {
            "Customer Name  " : self.customer_name,
            "Item ":self.items,
            "Delivery charge ":self.__delivery_charge,
            "Total Bill ":self.__final_amount()
        }

    def customer_view(self):
        return {
            "Customer Name ":self.customer_name,
            "Item ":self.items,
            "Total Bill ":self.__final_amount()
        }

class adminpanel:
    def admin_view(self,food_order):
        return food_order._admin_view()

class user_app:
    def user_view(self,food_order):
        return food_order.customer_view()

order = food_order("arul",["pizza","burger"],50,500)

admin = adminpanel()
user  = user_app()

print(admin.admin_view(order))
print(user.user_view(order))
```

```python
class BankAccount:
    def __init__(self, name, amount):
        self.__account_holder = name
        self.__balance = amount

    def deposit_amount(self, deposit):
        self.__balance += deposit
        return f"Deposited Amount: {deposit}"

    def withdrawal_amount(self, withdrawal):
        if withdrawal > self.__balance:
            return "Insufficient Balance"
        self.__balance -= withdrawal
        return f"Withdrawn Amount: {withdrawal}"

    def details(self):
        return {
            "Name": self.__account_holder,
            "Balance": self.__balance
        }


deposit = int(input("Enter the deposit amount: "))
withdrawal = int(input("Enter the withdrawal amount: "))

b = BankAccount("Arul", 1000)

print("Before Transaction:", b.details())
print(b.deposit_amount(deposit))
print(b.withdrawal_amount(withdrawal))
print("After Transaction:", b.details())
```
